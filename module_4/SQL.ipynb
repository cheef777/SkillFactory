Вот на эти рейсы можно рекомендовать присмотреться повнимательнее. Ярко выраженной зависимости от даты вылета не наблюдается. Возможно это своего рода выбросы.

```python
def q(s):
    IQR = s.quantile(0.75) - s.quantile(0.25)
    perc25 = s.quantile(0.25)
    perc75 = s.quantile(0.75)
    f = perc25 - 1.5*IQR
    l = perc75 + 1.5*IQR
    print(
        '25-й перцентиль: {},'.format(perc25),
        '75-й перцентиль: {},'.format(perc75),
        "IQR: {}, ".format(IQR),
        "Границы выбросов: [{}, {}].".format(f, l))
    return perc25, perc75, f, l
```

```python
q(df.occupancy)
```
